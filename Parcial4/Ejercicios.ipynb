```python
import numpy as np
from scipy.optimize import minimize

def restriccion1(xyz):
    x,y,z = xyz
    return 2*x-4*y+5*z-2

def funcion1(xyz):
    x,y,z = xyz
    return x**2+y**2+z**2-2*z+1

inc = [1,1,1]

const1 = {
    'type': 'eq',
    'fun': restriccion1
}

result = minimize(funcion1, inc, constraints=const1)

print(result.fun)
```

La caja tiene cinco tapas, dos en el plano yz, dos en el plano xz y una en el plano xy. Si el material total es de $12 cm^2$ entonces xy + 2yz + 2xz = 12. Por otra parte, el volumen de la caja es sencillamente xyz.

```python
def restriccion2(xyz):
    x,y,z = xyz
    return x*y + 2*y*z + 2*x*z - 12

def funcion2(xyz):
    x,y,z = xyz
    return -x*y*z

inc = [1,1,1]

const2 = {
    'type': 'eq',
    'fun': restriccion2
}

result2 = minimize(funcion2, inc, constraints=const2)

print(abs(result2.fun))
```

El volumen máximo es de $ 4 cm^3$.

Por multiplicadores de Lagrange:

$L = XYZ + \lambda(xy+2yz+2xz - 12)$

$\frac{dL}{dx} = YZ + \lambda y + 2 \lambda z = 0$

$\frac{dL}{dy} = XZ + \lambda x + 2 \lambda z = 0$

$\frac{dL}{dz} = XY + 2 \lambda y + 2 \lambda x = 0$

$\frac{dL}{d \lambda} = xy + 2yz + 2xz - 12 = 0$

Solucionando el sistema se obtiene:

$x = 2, y = 2, z = 1$

Con lo cual

$A = XYZ = 2 \times 2 \times 1 = 4$

4.

Conviene calcularlo en primer lugar a la inversa, es decir, la probabilidad de que en un grupo de n personas no se repita un cumpleaños. Esto corresponde a una permutación simple, como el día tiene 365, entonces la probabilidad viene siendo $\frac{365}{365} * \frac{364}{365} * \frac{363}{365} ... \frac{365-n}{365}$. Cuya fórmula general sería:

$ p = \frac{365!}{365^n (365-n)!}$

Siendo evidente que si n es mayor a 365, la probabilidad es cero.

El complemento de que de un grupo de n personas, ninguna repita cumpleaños. Es que al menos haya un cumpleaños repetido. Con lo cuál la probabilidad que pide el ejercicio se puede calcular completamente como:

$p = 1 - \frac{365!}{365^n (365-n)!}$



```python
import math
import matplotlib.pyplot as plt

n = np.arange(1,81,dtype=int)
p = []

for i in range(1,81):
    p.append(1 -math.factorial(365)/ (365**i * math.factorial(365- i)))

plt.figure(dpi=100)
plt.plot(n,p)
plt.title("Probabilidad de cumpleaños repetido")
plt.xlabel("N personas")
plt.ylabel("Probabilidad P(N)")
plt.show()
```
